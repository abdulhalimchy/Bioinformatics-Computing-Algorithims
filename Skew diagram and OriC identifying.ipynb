```python
import matplotlib.pyplot as plt


dna = input("Dna String: ")  #CATGGGCATCGGCCATACGCC

x=[0]
y=[0]
oriCx=[]
oriCy=[]

cnt=0
mn=10000
mx=0
for i in range(0,len(dna)):
    if i!=0 and dna[i]=='G':
        if dna[i-1]=='C':
            oriCx.append(i)
            oriCy.append(cnt)
    if dna[i]=='G':
        cnt+=1
    elif dna[i]=='C':
        cnt-=1
    mn=min(cnt,mn)
    mx=max(cnt,mx)
    x.append(i+1)
    y.append(cnt)

    
# plotting the points
#plt.plot(x, y, color='green', linestyle='dashed', linewidth = 3,marker='o', markerfacecolor='blue', markersize=5)
#skew diagram
plt.figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(x, y)

#oric marking
#plt.scatter(oriCx, oriCy, marker='o', s=50, color='red')
for i in range(0,len(oriCx)):
    plt.scatter(oriCx[i], oriCy[i], marker='o', s=50, color='red')
    plt.annotate("OriC",(oriCx[i]-0.5, oriCy[i]-0.3))


#labeling x-axis and coloring
plt.xticks(range(len(dna)+1), " "+dna, color='black', weight='600')
for i in range(0,len(dna)):
    if dna[i]=='G':
        plt.gca().get_xticklabels()[i+1].set_color("green")
    elif dna[i]=='C':
        plt.gca().get_xticklabels()[i+1].set_color("red")
        
# setting x and y axis range 
plt.xlim(0,len(dna))
plt.ylim(mn-1, mx+1)


plt.xlabel('DNA Sequence')
plt.ylabel('#G-#C')
plt.title('Skew Diagram')

# function to show the plot
#plt.savefig('skew oric')
plt.show()
```
